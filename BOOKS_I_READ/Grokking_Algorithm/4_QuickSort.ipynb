Quick Sort is a major divide and conquer algorithm

## QuickSort Algorithm

```python
def quicksort(array):
    if len(array) < 2:
        return array       #base case : arrays with length 1 are already sorted
    else:
        pivot = array[0]
        less = [i for i in array[1:] if i <= pivot]       #defining subarray to left
        greater = [ i for i in array[1:] if i > pivot]    #defining subarray to right
        
        return (quicksort(less) + [pivot] + quicksort(greater))
    
    
quicksort([5,10,4,26,12,18])
```

In worst case , quicksort takes $O(n^2)$ and best case is $0(log\ n)$<br>
Quicksort is one of the fastest sorting algorith amomg all

## Sleep

```python
from time import sleep
def print_items2(list):
    for item in list:
        sleep(1)
        print (item)
        
print_items2([1,3,7,3,89])
```

Even though the time complexity is $O(n)$, we take a **sleep** of 1 second , so we would write the time complexity as $c * n$ where $c$ is some constant time

These constant usually don't make much difference. But sometimes, they do.

## One such example is difference between *quick sort* and *merge sort*<br>
Quicksort has a smaller constant than merge sort. So if theyre both $O(n * log n)$ time, quicksort is faster. And quicksort is faster in practice because it hits the average case way more oten than the worst case.
