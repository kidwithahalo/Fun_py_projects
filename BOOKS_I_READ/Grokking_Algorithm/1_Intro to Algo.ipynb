#### Algorithm is a set of instructions for carrying out/accomplishing any task.

In general, for any list of $n$, binary search will take $log_2 n$ steps to run in the worst case, whereas simple search will take $n$ steps.

### Binary Search

```python
def binary_search(list, item):
    low = 0
    high = len(list) -1
    while low <= high:
        mid = (low + high)
        guess = list[mid]
        if guess == item:
            return mid
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None
```

```python
my_list = [1,2,6,8,5,2,7,5,8,3]
binary_search(my_list, 2)
# returns index
```

<font color='red'> Big $O$ notation is special notation that tells you how fast an algorithm is. It is not in terms of time, it tells us how the algorithm grows with time. <br> It is the count of operations algorithm will make

### <font color='magenta'> Some Big $O$ times
• $O(log n)$, also known as log time. Example: Binary search.<br>
• $O(n)$, also known as linear time. Example: Simple search.<br>
• $O(n * log n)$. Example: A fast sorting algorithm, like quicksort.<br>
• $O(n^2)$. Example: A slow sorting algorithm, like selection sort <br>
• $O(n!)$. Example: A really slow algorithm, like the traveling salesperson
