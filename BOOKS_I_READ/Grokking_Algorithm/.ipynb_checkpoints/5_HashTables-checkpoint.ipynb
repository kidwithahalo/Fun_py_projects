### Voter's credibility using Hash

```python
voted = {}

def check_voter(name):
    if voted.get(name):
        print('Bhagao mc ko!')
    else:
        voted[name] = True
        print("Aap aaiye!")
        
        
check_voter("mahindera")
check_voter('mahesh')
check_voter('mahesh')

# Checking duplicates is quite fast with hash tables
```
