Logistic function is defined as $$\frac{1}{1 + e^{-x}}$$

```python
import math

import matplotlib.pyplot as plt
import numpy as np

def logistic(x:float)->float:
    return 1.0/(1+math.exp(-x))

xs = np.linspace(-10,10)
ys = [logistic(i) for i in xs]

plt.plot(xs,ys)
plt.show()
```

```python
def logistic_prime(x:float) -> float:
    y = logistic(x)
    return y * (1 - y)

xs = np.linspace(-10,10)
ys = [logistic_prime(i) for i in xs]

plt.plot(xs,ys)
plt.show()
```
