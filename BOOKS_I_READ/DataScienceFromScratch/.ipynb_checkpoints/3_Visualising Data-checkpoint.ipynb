## <font color = 'aqua'>Line Charts

```python
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8,5)
matplotlib.rcParams['figure.dpi'] = 150
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# creating a line chart
# fig = plt.figure(figsize=(3,2),dpi = 200)
plt.plot(years, gdp, color= 'g', marker = 'o', linestyle = 'solid')

plt.title("Nominal GDP")
plt.xlabel("Years")
plt.ylabel("Billions of $")
plt.show()
```

## Bar Charts 

```python
movies = ['Annie Hall', 'Ben-hur', 'Casablanca','Gandhi','West Side Story']
num_oscar = [5,11,3,8,10]

# plot bars with left x-Coordinates 
ax = plt.bar(range(len(movies)), num_oscar, color = 'teal')

plt.title("My favourite Movie")
plt.ylabel("# of Academy Awards")

# label x-axis with movie names at center
plt.xticks(range(len(movies)), movies)
    

plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# bucketing grades as per decile
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()],    #shift bars right by 5
       histogram.values(),                  #give each bar its correct height
       10,                                  #each bar given width of 10
       edgecolor = (0,0,0))

plt.axis([-5,105,0,5])                     #x axis form -5, 105 and y-axis from 0 , 5


plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("Number of students")
plt.title("Distribution of marks")
plt.tight_layout()
plt.show()
```

# <font color = 'cyan'> Line Charts

```python
import numpy as np

variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]

total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]

# making multiple calls to plot
plt.plot(xs, variance, 'g:', label = 'Variance')
plt.plot(xs, bias_squared, 'r-', label= r"$Bias^2$")
plt.plot(xs, total_error, 'b-.', label = "Total Error")
plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks(xs)
plt.title('The Bias Variance Tradeoff')

plt.axhline(np.min(total_error))
plt.text(4, 50, s= 'Best Tradeoff = '+str(np.min(total_error)), fontsize=10, va='center', ha='center', backgroundcolor='w',
         bbox = dict(edgecolor = 'red'))


plt.show()
```

#### Some more things to take care of 

```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)

# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
plt.gca().clear()
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
plt.ticklabel_format(useOffset=False)

plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not So Huge Anymore")
```

# ScatterPlot

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']

plt.scatter(friends,minutes)

# labelling each point - annotation

for label, friend_count,minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy = (friend_count,minute_count),
                 xytext = (5,-5),
                 textcoords = 'offset points'
                )
    
plt.title('Daily minutes vs Number of Friends')
plt.xlabel("Number of Friends")
plt.ylabel('daily minutes spent on site')


plt.show()
```
