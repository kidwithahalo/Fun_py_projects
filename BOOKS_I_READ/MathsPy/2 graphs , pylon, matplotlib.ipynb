```python
'''
the relationship between gravitational force and
distance between two bodies 
'''

import matplotlib.pyplot as plt

# draw the graph
def graph(x,y):
    plt.plot(x,y, marker = 'o')
    plt.xlabel('Distance in metres')
    plt.ylabel('G in N')
    plt.show()
    
def F_r():
#     generating values for R
    r = range(100,1001,50)
    F = []
    
    
#     constant of G
    G = 6.674 * (10 ** -11)
    m1 = 0.5
    m2 = 1.5
    
#     calculating force 
    for dist in r:
        force = (G*(m1*m2))/(dist**2)
        F.append(force)
        
    graph(r,F)
    
if __name__ == '__main__':
    F_r()
```

For this challenge, youll write a program that creates a bar chart for easy comparison of
weekly expenditures. The program should first ask for the number of categories for 
the expenditures and the weekly total expen- diture in each category, and then it
should create the bar chart showing these expenditures.

```python
# Visualising your expenses

import matplotlib.pyplot as plt

def create_bar_chart(data, labels):
    
    num_bars = len(data)
    positions = range(1, num_bars + 1)
#     this is the list of point on y axis where each bar is centered

    plt.barh(positions, data, align = "center" , color = 'g')
    
    plt.yticks(positions, labels)
    # Get or set the current tick locations and labels of the y-axis and titling as well

    plt.xlabel("Amount")
    plt.ylabel("Categories")
    plt.title("Weekly expenditures")
    
#     turnd on thr grid which may assisy in visual estimation
    plt.grid()
    plt.show()
    
if __name__ == "__main__":
    
    n = int(input("Enter the number of categories : "))
    labels = []
    expenditures = []
    for i in range(n):
        category = input("Enter category : ")
        expense = float(input("Expenditure : "))
        
        labels.append(category)
        expenditures.append(expense)
        
    create_bar_chart(expenditures, labels)
        
    
    

```

```python
import matplotlib.pyplot as plt

def fibo(n):
    
    if n == 1:
        return [1]
    elif n == 2:
        return [1,1]
    
    a = 1
    b = 1
    series = [a,b]
    for i in range(n):
        c = a + b
        series.append(c)
        a = b 
        b = c
        
    return series

def plot_ratios(series):
    ratios = []
    
    for i in range(len(series)-1):
        ratios.append(series[i+1]/series[i])
        
    
    plt.title("Ratio btw Fibonacci numbers and Golden ratio")
    plt.ylabel("Ratio")
    plt.xlabel("Num")
    plt.plot(ratios)
    plt.show()
    
    
if __name__ == "__main__":
    
#     number of fibonacci numbers
    num = 100
    series = fibo(num)
    plot_ratio(series)
```
