```python
# Starting with basics 

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],'r*') #markersize = 20

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First Plot')
plt.show()
```

```python
# Alpha 

import matplotlib.pyplot as plt

plt.plot([1,2,3],[4,5,6], linewidth=20, zorder= 2)
plt.plot([1,2,5,9,10], linewidth=20, zorder  = 1)

#alpha renders the graph as transparent, default is 1
#xlabel(xlabel,fontdic=None, labelpad=None, kwargs)
#fontdic is used as {"color":"red"} if to change colors
#backgroundcolor="red" renders a bg to axes labelpad
#fontsize or font
#fontstyle or style
#font-weight or weight
#fontfamily or family

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("First Plot")
#same alterations with title too
#loc attribute in title is alignment . left right or center
plt.show()
```

```python
# Bargraph

#bar(x,height,width=0.8, bottom = None, align = 'center', data = None, **kwargs)

import matplotlib.pyplot as plt
x = ['science', 'arts', 'commerce']
h = [200,300,100]
c = ['red', 'blue', 'green']
plt.bar(x,h, bottom = 100, color = c)
#edgecolor
plt.xlabel('courses')
plt.ylabel('students enrolled')
plt.show()
```

```python
# Gridlines

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8],'r*')
#markersize = 20

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First Plot')
plt.grid(b=True)
#grid(b,which='major',axis='both',**kwargs)
plt.show()
```

```python
# Horizontal Graphs

#barh(y,width, left = None , align , **kwargs)

import matplotlib.pyplot as plt
y = ['science', 'arts', 'commerce']
w = [200,300,100]
c = ['red', 'blue', 'green']
plt.barh(y,w,color = c)
#edgecolor
plt.xlabel('courses')
plt.ylabel('students enrolled')
plt.show()
```

```python
# Legends in Graphs

import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

plt.plot(t**2, t, color="red", label="square")
plt.plot(t**3, t, color="blue", label="cubes")
#labels goes as the nomenclature of legend
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First Plot')
plt.legend(loc="lower right")
#ncol=2, the legends appears 2 at a time
#legend is important while graphing
plt.show()
```

```python
# Limits

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])

left,right = plt.xlim()
#checks axis range - also we can limit the range of axis
#xlim(left,right)

print(left,right)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First Plot')
plt.show()
```

```python
# Markers in graph

import matplotlib.pyplot as plt

plt.plot([10,20,30,40], marker="+" )

#linewidth=1
#dashes=[2,10] as 2 is dashses and 10 is distance between them
#marker = "+" appears as marker on graph
#markeredgecolor="green"
#markeredgewidth="20" increases the thickness of marker
#markersize="20" increase the size of marker
#markerfacecolor="red"
#markevery=2 , will only point every second markings.

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("First Plot")
plt.show()
```

# Numpy 

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0,10,0.5)
plt.plot(a,'r*', a**3,'b^',a**2,'co:') #markersize = 20
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('First Plot')
plt.show()
```

```python
# Sine function 

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi)

plt.plot(x,np.sin(x), marker = '^', label = 'sin')
plt.plot(x,np.cos(x), marker = '*', label = 'cos')
plt.legend()
plt.text(2,20,"Hello")
#x, y , text . usual cordinates
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = [0,30,45,60,90]

#to convert degrees into radians
x = [i*(np.pi/180) for i in t]

plt.plot(x,np.sin(x), marker = '^', label = 'sin')
plt.plot(x,np.cos(x), marker = '*', label = 'cos')
plt.legend()



plt.show()
```

```python
# Ticks

import matplotlib.pyplot as plt
import numpy as np

t = np.array([1,2,3,4])

plt.plot(t**2 , t, color = 'red', label = 'label1')
plt.plot(t**3 , t, color = 'green', label = 'label2')

plt.xlim(0,5)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('First Plot')
plt.grid()
plt.xticks([2,4,5,7],['A=2','B=4','C=5','D=7'])
plt.yticks(np.arange(1,21,step = 4))
#nubmer from  1 to 21 with jump f 4
#ticks are useful when giving values name in axis

plt.tick_params(width = 5, color = 'red', direction = 'inout')
#tick_params(axis='x,y,both', **kwargs)
# kwargs = length, width, color , pad

plt.show()
```
